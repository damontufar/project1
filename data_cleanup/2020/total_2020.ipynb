```python
#Import dependencies
import pandas as pd
import numpy as np
import os
```

```python
base_path= 'Output/'
denue_df = pd.read_csv(f'{base_path}denue_total.csv')
pib_df = pd.read_csv(f'{base_path}pib.csv')
```

```python
denue_df.head()
```

```python
pib_df.head()
```

```python
column_names = ["CODIGO", "DELTA PIB","ID_ESTRATO"]
pib2_df = pd.DataFrame(columns =  column_names)

for cod, delta in pib_df[["CODIGO", "DELTA PIB"]].values:
    df2 = pd.DataFrame([[cod, delta, "0 a 5 personas"], [cod, delta, "6 a 10 personas"]], columns = column_names)
    pib2_df = pib2_df.append(df2, ignore_index = True)

pib2_df
```

```python
merge_df = pd.merge(denue_df, pib_df, on="CODIGO")
merge_df
```

```python
final_df = merge_df[["CVE_ENT","CODIGO","Concepto","ID_ESTRATO", "Unidades 2020", "DELTA UNIDADES_ECONOMICAS", "DELTA PIB"]]

final_df
```

```python
#Export the dataframe to a csv file
csv_path= "Output/denue_total.csv"
merge_df.to_csv(csv_path,index=False)
```
