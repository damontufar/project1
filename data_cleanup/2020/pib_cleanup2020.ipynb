```python
#Import dependencies
import pandas as pd
import numpy as np
import os
```

```python
# Read the CSV file from EMEC
base_path= "Resources/pib/PIBT_2.xlsx"
reader_df = pd.read_excel(base_path)
reader_df.head()
```

```python
reader_df['Concepto'] =  reader_df['Concepto'].str.strip()
```

```python
reader_df['Concepto'].values
```

```python
general_df = reader_df[(reader_df["Concepto"].str.match('^\d{2}\s\-'))]
general_df["CODIGO"] = general_df["Concepto"].str[:2]
general_df.head()
```

```python
general_df = general_df[["Concepto", "CODIGO", "2019 - 9 Meses", "2020 - 9 Meses"]]
general_df["DELTA PIB"] = ((general_df["2020 - 9 Meses"]/general_df["2019 - 9 Meses"])-1)*100
general_df
```

```python
#Export the dataframe to a csv file
csv_path= "Output/pib.csv"
general_df.to_csv(csv_path,index=False)
```
