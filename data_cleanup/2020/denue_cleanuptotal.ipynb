```python
#Import dependencies
import pandas as pd
import numpy as np
import os
```

```python
base_path= 'Output/'
denue20_df = pd.read_csv(f'{base_path}denue_2020.csv')
denue19_df = pd.read_csv(f'{base_path}denue_2019.csv')
```

```python
denue20_df.head()
```

```python
denue19_df.head()
```

```python
merge_df = pd.merge(denue19_df, denue20_df, on=["CVE_ENT","CODIGO","ID_ESTRATO"])
merge_df = merge_df.rename(columns = {"Unidades_Economicas_x": "Unidades 2019", "Unidades_Economicas_y": "Unidades 2020"})
```

```python
merge_df["DELTA UNIDADES_ECONOMICAS"] = merge_df["Unidades 2020"] - merge_df["Unidades 2019"]
merge_df
```

```python
#Export the dataframe to a csv file
csv_path= "Output/denue_total.csv"
merge_df.to_csv(csv_path,index=True)
```
