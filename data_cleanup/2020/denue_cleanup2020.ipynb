```python
#Import dependencies
import pandas as pd
import numpy as np
import os
```

```python
# Read the CSV files that will be analyzed from the DENUE site
base_path= "Resources/denue"
file_list = os.listdir(base_path)
denue_columns = ["codigo_act", "nombre_act","per_ocu","cve_ent","entidad"]
denue_df = pd.DataFrame(columns = denue_columns)
total = 0

for file in file_list:
    reader_df = pd.read_csv(f'{base_path}/{file}', encoding="latin-1")
    clean_df = reader_df[denue_columns]
    clean_df = clean_df[(clean_df['per_ocu'] == "0 a 5 personas") | (clean_df['per_ocu'] == "6 a 10 personas")]
    denue_df = denue_df.append(clean_df, ignore_index = True)
    a = len(clean_df)
    total = total + a

print(total)
```

```python
denue_df['cve_ent'] = denue_df['cve_ent'].astype(str)
denue_df['cve_ent'] = denue_df['cve_ent'].apply(lambda x: x.zfill(2))
```

```python
denue_df["codigo_act"] = denue_df["codigo_act"].astype(str)
denue_df["CODIGO"] = denue_df["codigo_act"].str[:2]
denue_df.columns = ["COD_ACT","NOMBRE_ACT", "ID_ESTRATO","CVE_ENT","ENTIDAD","CODIGO"]
denue_df
```

```python
grp_df = denue_df.groupby(["CVE_ENT", "CODIGO", "ID_ESTRATO"])\
        .agg({"ID_ESTRATO":"count"})\
        .rename(columns = {"ID_ESTRATO": "Unidades_Economicas"})
grp_df
```

```python
#Export the dataframe to a csv file
csv_path= "Output/denue_2020.csv"
grp_df.to_csv(csv_path,index=True)
```
